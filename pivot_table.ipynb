```python
corrs = []
for i in range(1000):
    a = np.random.randn(10000)
    b = np.random.randn(10000)
    corrs.append(np.corrcoef(a, b)[0, 1])
```

```python
corrs = np.array(corrs)
```

```python
sns.histplot(corrs)
```
